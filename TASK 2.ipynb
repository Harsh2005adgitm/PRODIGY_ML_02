```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("Retail_Customers_200.csv")
file
```

```python
features = file [["Age", "Annual Income (k$)", "Spending Score (1-100)" ]]
```

```python
scaler = StandardScaler() 
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=5, random_state=222, n_init=10)
file['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(20, 6))
plt.scatter(file['Annual Income (k$)'], file['Spending Score (1-100)'], c=file['Cluster'], cmap='viridis', s=50)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments using K-means Clustering')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```
